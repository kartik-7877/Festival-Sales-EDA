### Exploratory Data Analysis¶

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Gender Distribution')
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot( x='Gender', y='Amount', data= sales_gen,hue='Gender')
ax.set_title('Gender Wise Sales Comparision');
```

### From above graph we can see that most of the buyers are females and even the purchanging power of female are grater than male.

### Age base insights

```python
df.columns
```

```python
ax = sns.countplot(data=df, x='Age Group',hue='Gender')
for bars in ax.containers: 
    ax.bar_label(bars)
ax.set_title('Age Group Distribution by Gender')
```

```python
# Total amount vs Age group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot( x='Age Group', y='Amount', data= sales_age, hue='Age Group');
```

### from above graph we can see that most of the buyers are of age group is 26 to 35

### State base insights

```python
df.columns
```

```python
# Total numbers of orders from top 10 states
sales_states = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax = sns.barplot(data=sales_states, x='State', y='Orders', hue='State')
sns.set( rc={'figure.figsize':(17,5)})
ax.set_title('Total Orders From Top 10 States');
```

```python
df.columns
```

```python
# Total amount vs from top 10 states
total_amount_by_states = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(data=total_amount_by_states, x='State', y='Amount', hue='State')
ax.set_title('Total amount vs from the top 10 State');
```

### Marital base Insights

```python
ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status')
sns.set(rc={'figure.figsize':(5,3)})
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Number of Marital Status');
```

```python
sales_by_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,3)})
ax = sns.barplot(data=sales_by_status, x='Marital_Status', y='Amount', hue='Gender');
```

### From above graph we can see the most of the buyers are married (woman) and they have high purchasing power.

### Occupation base insights

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df, x='Occupation', hue='Occupation')

for bars in ax.containers: 
    ax.bar_label(bars)
ax.set_title('Distribution of People Across Occupations')
```

### From the above graph we can see form the IT sector more peaple puchasing.


```python
sales_occup = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.barplot(data=sales_occup, x='Occupation', y='Amount', hue='Occupation')
for bars in ax.containers: 
    ax.bar_label(bars)
ax.set_title('Occupation-Wise Sales Analysis')
```

#### From above graph we can see that most of buyers are working in IT, Aviation and helthcare center.

#### Product Catagory Analysis.

```python
most_selling_cat = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.barplot(data=most_selling_cat, x='Product_Category', y='Amount', hue='Product_Category')
ax.set_title('Most Selling Product Categories by Total Sales');
```

#### From above graphs we can see the most selling category is the food.

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category')
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
top_selling_product = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=top_selling_product, x='Product_ID', y='Orders', hue='Orders');
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar');
```
